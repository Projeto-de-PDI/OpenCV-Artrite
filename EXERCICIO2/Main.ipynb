```python
import cv2
import mediapipe as mp
import numpy as np
import joblib
```

```python
##Carregar modelo SaLVO
model = joblib.load('Modelo_Exercicio2.pkl')
```

```python
##Configurar MediaPipe
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
```

```python
cap = cv2.VideoCapture('DATA_SAMPLE/DATA7.mp4')
#cap = cv2.VideoCapture(0)
gestures_sequence = []

with mp_hands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5) as hands:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        #Processamento de imagem
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        results = hands.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                #Landmarks das mos
                landmarks = [landmark for landmark in hand_landmarks.landmark]
                flattened = [coord for landmark in landmarks for coord in (landmark.x, landmark.y, landmark.z)]

                #Predio com o modelo.
                prediction = model.predict([flattened])[0]
                gesture = prediction


                # Adicionar gesto  sequncia, mas apenas se for diferente do ltimo
                if not gestures_sequence or gestures_sequence[-1] != gesture:
                    gestures_sequence.append(gesture)
                
                # Manter a lista com no mximo 3 elementos
                if len(gestures_sequence) > 4:
                    gestures_sequence.pop(0)

                # Verificar se a sequncia corresponde ao "Exerccio 1"
                if gestures_sequence == [1, 2, 3, 2]:
                    cv2.putText(image, "Exercicio 1", (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
                    #gestures_sequence = []

                #Desenho e Texto
                mp_drawing.draw_landmarks(image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
                cv2.putText(image, str(prediction), (10, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

        cv2.imshow("Exercicio", image)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
